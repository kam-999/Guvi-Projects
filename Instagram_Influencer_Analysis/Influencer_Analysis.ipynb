```python
import pandas as pd
```

```python
df = pd.read_csv("Influencer.csv")
df
```

```python
df.info()
```

```python
#Convert objects with numerical values(Likes and Posts) into float

def convert_likes_to_float(likes_str):
    if 'b' in likes_str:
        return float(likes_str.replace('b', '')) * 1_000_000_000
    elif 'm' in likes_str:
        return float(likes_str.replace('m', '')) * 1_000_000
    elif 'k' in likes_str:
        return float(likes_str.replace('k', '')) * 1_000
    else:
        return float(likes_str)
    

df['Followers'] = df['Followers'].apply(convert_likes_to_float)
df['Avg. Likes'] = df['Avg. Likes'].apply(convert_likes_to_float)
df['New Post Avg. Likes'] = df['New Post Avg. Likes'].apply(convert_likes_to_float)
df['Posts'] = df['Posts'].apply(convert_likes_to_float)
df['Total Likes'] = df['Total Likes'].apply(convert_likes_to_float)
```

```python
#One-hot encode "Country Or origin" column

encoded_df = pd.get_dummies(df['Country Or Region'])

df_copy = df.copy()  

df_copy = pd.concat([df_copy, encoded_df], axis=1)

df_copy
```

```python
df_copy = df_copy.drop(df_copy.columns[[0,1,9]], axis=1)

df_copy
```

1. Highly correlated pairs of features and their correlation coefficients.

```python

correlation_matrix = df_copy.corr()

# Display the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
# Find pairs of features with a high correlation
high_corr_pairs = correlation_matrix.unstack().sort_values(ascending=False)
high_corr_pairs = high_corr_pairs[high_corr_pairs != 1]  # Exclude self-correlation
high_corr_pairs = high_corr_pairs[high_corr_pairs.abs() > 0.8]  # Adjust threshold as needed

high_corr_pairs = high_corr_pairs[high_corr_pairs.index.get_level_values(0) < high_corr_pairs.index.get_level_values(1)]

print(high_corr_pairs)
```

2. Frequency distribution of features

```python
import matplotlib.pyplot as plt


# Define the numerical features you want to analyze
features = ['Posts', 'Followers', 'Influence Score']

# Create histograms for each feature
for feature in features:
    plt.figure(figsize=(10, 6))
    plt.hist(df[feature], bins=30, edgecolor='k', alpha=0.7)
    plt.title(f'Frequency Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

3. Country with highest number of instagram influencers

```python
import seaborn as sns

country_counts = df['Country Or Region'].value_counts()

# Create a bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x=country_counts.index, y=country_counts.values, palette='viridis')

plt.title('Number of Instagram Influencers by Country')
plt.xlabel('Country')
plt.ylabel('Number of Influencers')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

# Display the country with the highest number of influencers
max_country = country_counts.idxmax()
max_count = country_counts.max()
print(f'The country with the highest number of Instagram influencers is {max_country} with {max_count} influencers.')
```

4. Top 10 Influencers based on Followers, Average Likes, Total Likes.

```python
top_10_followers = df.nlargest(10, 'Followers')

# Get the top 10 influencers based on Average Likes
top_10_avg_likes = df.nlargest(10, 'Avg. Likes')

# Get the top 10 influencers based on Total Likes
top_10_total_likes = df.nlargest(10, 'Total Likes')


# Set up the figure and axes
fig, axs = plt.subplots(3, 1, figsize=(12, 18))

# Plot Top 10 Influencers by Followers
sns.barplot(x='Followers', y='Channel Info', data=top_10_followers, ax=axs[0], palette='viridis')
axs[0].set_title('Top 10 Influencers by Followers')
axs[0].set_xlabel('Followers')
axs[0].set_ylabel('Influencer')

# Plot Top 10 Influencers by Average Likes
sns.barplot(x='Avg. Likes', y='Channel Info', data=top_10_avg_likes, ax=axs[1], palette='magma')
axs[1].set_title('Top 10 Influencers by Average Likes')
axs[1].set_xlabel('Average Likes')
axs[1].set_ylabel('Influencer')

# Plot Top 10 Influencers by Total Likes
sns.barplot(x='Total Likes', y='Channel Info', data=top_10_total_likes, ax=axs[2], palette='cividis')
axs[2].set_title('Top 10 Influencers by Total Likes')
axs[2].set_xlabel('Total Likes')
axs[2].set_ylabel('Influencer')

# Adjust layout
plt.tight_layout()
plt.show()
```

5. Relationships between features represented by scatter plots

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scale the features
df['Followers (Millions)'] = df['Followers'] / 1e6
df['Total Likes (Millions)'] = df['Total Likes'] / 1e6
df['Average Likes (Millions)'] = df['Avg. Likes'] / 1e6
df['Posts (Thousands)'] = df['Posts'] / 1e3

# Set up the figure and axes
fig, axs = plt.subplots(2, 2, figsize=(14, 12))


# Scatter plot: Followers (Millions) vs. Total Likes (Millions)
sns.scatterplot(x='Followers (Millions)', y='Total Likes (Millions)', data=df, ax=axs[0, 0], color='b')
axs[0, 0].set_title('Followers vs. Total Likes (Millions)')
axs[0, 0].set_xlabel('Followers (Millions)')
axs[0, 0].set_ylabel('Total Likes (Millions)')
axs[0, 0].grid(True)

# Scatter plot: Followers (Millions) vs. Influence Score
sns.scatterplot(x='Followers (Millions)', y='Influence Score', data=df, ax=axs[0, 1], color='g')
axs[0, 1].set_title('Followers vs. Influence Score')
axs[0, 1].set_xlabel('Followers (Millions)')
axs[0, 1].set_ylabel('Influence Score')
axs[0, 1].grid(True)

# Scatter plot: Posts (Thousands) vs. Average Likes (Millions)
sns.scatterplot(x='Posts (Thousands)', y='Average Likes (Millions)', data=df, ax=axs[1, 0], color='r')
axs[1, 0].set_title('Posts vs. Average Likes (Millions)')
axs[1, 0].set_xlabel('Posts (Thousands)')
axs[1, 0].set_ylabel('Average Likes (Millions)')
axs[1, 0].grid(True)

# Scatter plot: Posts (Thousands) vs. Influence Score
sns.scatterplot(x='Posts (Thousands)', y='Influence Score', data=df, ax=axs[1, 1], color='purple')
axs[1, 1].set_title('Posts vs. Influence Score')
axs[1, 1].set_xlabel('Posts (Thousands)')
axs[1, 1].set_ylabel('Influence Score')
axs[1, 1].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```
